```python
def categorize_success(rating):
    if rating >= 9:
        return 'Excellent'
    if rating >= 8:
        return 'Very Good'
    if rating >= 7:
        return 'Good'
    if rating >= 6:
        return 'Above Average'
    if rating >= 5:
        return 'Average'
    if rating >= 4:
        return 'Below Average'
    if rating >= 3:
        return 'Bad'
    if rating >= 2:
        return 'Very Bad'
    if rating >= 1:
        return 'Disasterous'
    if rating>=0:
        return 'Worst'
    
df["rating_category"] = df["vote_average"].apply(categorize_success)
```
